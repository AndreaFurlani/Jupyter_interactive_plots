```python
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import io
from bokeh.io import show, output_notebook
from bokeh.models import ColorBar, ColumnDataSource, CategoricalColorMapper
from bokeh.plotting import figure
from bokeh.transform import transform
import bokeh.palettes

output_notebook(hide_banner=True)
```

```python
EXPLAINATION = """\
<div class="app-sidebar">

<p>Load a .cvs, .txt, .dat file containing the counts of phenotypes for each gene contained in an <a href="https://www.mousephenotype.org/help/data-visualization/gene-pages/phenogrid/">
IMPC major category</a>. Strating from a list of MGI genes ids or Symbols, that kind of matix can be obtained here. Using the slider, adjust the treshold to set the minimum count to be displayed in the heatmap.</p>
</div>
"""
```

```python
class App:
    def __init__(self):
        self._plot_container = widgets.Output()
        self._threshold_slider, slider_box = self._create_slider()
        self._load_button, load_box = self._create_load_button()
        
        _app_container = widgets.VBox([
            widgets.VBox([load_box,slider_box])], layout = widgets.Layout(align_items='center', flex='3 0 auto'))
        
        self.container = widgets.VBox([
            widgets.HTML(('<h1><em><b>Gene vs IMPC phenotype category heatmap</b></em></h1>'), 
                         layout=widgets.Layout(margin='0 0 5em 0')), 
            widgets.HBox([
                _app_container, 
                widgets.HTML(EXPLAINATION, layout=widgets.Layout(margin='0 0 0 2em'))
            ], layout= widgets.Layout(flex='1 1 auto', margin='0 auto 0 auto', max_width='1024px')),
        self._plot_container
        ], layout= widgets.Layout(align_items='center', flex='1 1 auto', margin='0 auto 0 auto'))
        
    
    def _create_load_button(self):
        load_label = widgets.Label('File to load:')
        load_b = widgets.FileUpload(accept=".txt, .csv, .dat", multiple=False)
        load_b.observe(self._on_change_l, names=['value'])
        load_box = widgets.HBox([load_label,load_b])
        return (load_b, load_box)
        
        
    def _on_change_l(self, _):
        inp = list(self._load_button.value.values())[0] #if multiple setted to true in FileUpload(), will not work!
        content = inp['content']
        content = io.StringIO(content.decode('utf-8'))
        self.mat = pd.read_csv(content, sep="\t", index_col=0)
        self.mat.index.name = 'MGI_id'
        self.mat.columns.name = 'phen_sys'
        self._update_app()
        
    
    def _create_slider(self):
        slider_label = widgets.Label('Threshold: ')
        slider = widgets.IntSlider(value=0, min=0, max = 0, step=1, orintation='horizontal', readout=True, readout_format="d")
        slider.observe(self._on_change, names=['value'])
        slider_box = widgets.HBox([slider_label,slider])
        return (slider, slider_box)
    
   
    def _on_change(self, _):
        self._update_app()
        
    
    def _update_app(self):
        self._threshold_slider.max = max(self.mat.max(numeric_only=True))
        threshold = self._threshold_slider.value
        self._plot_container.clear_output(wait=True)
        try:
            with self._plot_container:
                p = self._create_plot(threshold)
                show(p, notebook_handle=True)
        except (NameError,AttributeError) as e:
            pass 
            
    def _create_plot(self, threshold):
        
        tem_mat= self.mat.copy()
        rem=[]
        if threshold != 0:
            for i in tem_mat.index:
                if tem_mat.loc[i].max() < threshold:
                    rem.append(i)
            tem_mat.drop(rem,inplace=True,axis=0)

        #Create a custom palette and add a specific mapper to map color with values, we are converting them to strings to create a categorical color mapper to include only the
        #values that we have in the matrix and retrieve a better representation
        df = tem_mat.stack(dropna=False).rename("value").reset_index()
        fact= df.value.unique()
        fact.sort()
        fact = fact.astype(str)
        df.value = df.value.astype(str)

        mapper = CategoricalColorMapper(palette=bokeh.palettes.inferno(len(df.value.unique())), factors= fact, nan_color = 'gray')
        
        print(df.head(10))
        
        #Define a figure
        p = figure(
            plot_width=1280,
            plot_height=800,
            x_range=list(df.phen_sys.drop_duplicates()[::-1]),
            y_range=list(df.MGI_id.drop_duplicates()),
            tooltips=[('Phenotype system','@phen_sys'),('Gene','@MGI_id'),('Phenotypes','@value')],
            x_axis_location="above",
            output_backend="webgl",
            toolbar_location="right",
            tools="pan,wheel_zoom,box_zoom,reset,save")

        #Create rectangles for heatmap
        p.rect(
            x="phen_sys",
            y="MGI_id",
            width=1,
            height=1,
            source=ColumnDataSource(df),
            fill_color=transform('value', mapper))
        p.xaxis.major_label_orientation = 45

        #Add legend
        color_bar = ColorBar(
        color_mapper=mapper,
        label_standoff=6,
        border_line_color=None)
        p.add_layout(color_bar, 'right')
        return(p)
        
```

```python
app = App()
app.container
```
